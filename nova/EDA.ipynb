```python
data['home_country'] = (
    data['home_country']
    .str.strip()          # remove spaces
    .str.upper()          # make everything uppercase
)
```

```python
data['kyc_tier'] = data['kyc_tier'].replace({
    'standard ': 'STANDARD',
    'enhancd': 'ENHANCED',
    'nan'    :   'NAN'})
```

```python
# Drop identifier columns

# Handle duplicate records

# Fix home country inconsistencies

# Handle missing values

# Perform exploratory data analysis (EDA)

# Correct spelling issues
```

```python

# HANDLE DUPLICATE RECORDS

print("Duplicate rows:", data.duplicated().sum())

data = data.drop_duplicates()

print("Duplicates removed.")
```

```python
data.duplicated().sum()
```

```python
# DROP IDENTIFIER COLUMNS


data = data.drop(columns=['transaction_id','customer_id', 'device_id','ip_address'], errors='ignore')
```

```python
print(data.columns)
```

```python
# HANDLE MISSING VALUES
data.head(5)
```

```python
data['amount_src'] = data.groupby('source_currency')['amount_src'].transform(lambda x: x.fillna(x.median()))
```

```python
for col in data.select_dtypes(include=['object','string']):
    data[col] = data[col].fillna(data[col].mode()[0])
    print(col)
```
